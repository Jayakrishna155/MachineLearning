```python
data = pd.read_csv("drive/MyDrive/Datasets/Housing.csv",usecols=['area','bedrooms','parking','price'])
data.head()
data.fillna({'bedrooms':data.bedrooms.median()},inplace=True)
```

```python
model = linear_model.LinearRegression()
model.fit(data[['area','bedrooms','parking']],data.price)
```

```python
model.predict([[4000,3,2]])
```

```python
from sklearn.metrics import mean_squared_error,r2_score
y_true = data.price
y_pred = model.predict(data[['area','bedrooms','parking']])
# mean_squared_error(y_true,y_pred)
r2_score(y_true,y_pred)
```
