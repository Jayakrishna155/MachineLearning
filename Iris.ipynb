```python
from sklearn.datasets import load_iris
data = load_iris()
data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data.data,data.target,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.predict([[4.7, 3.2, 1.3, 2.2]])
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from seaborn import heatmap as hm
from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))
hm(cm,annot = True,)
```

```python
#SVM (Support Vector Machine)
from sklearn.datasets import load_iris
data = load_iris()
dir(data)
```

```python
import pandas as pd
df  = pd.DataFrame(data.data,columns=data.feature_names)
df.head()
```

```python
df['target'] = data.target
```

```python
df['target']
```

```python
df['flower'] = df.target.apply(lambda x:data.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10,10))
plt.scatter(df['petal length (cm)'],df['petal width (cm)'],c=df['target'])
```

```python
from sklearn.model_selection import train_test_split
x = df.drop(['target','flower'],axis='columns')
y = df.target
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
